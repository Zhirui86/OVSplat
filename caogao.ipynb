```python
import torch
import torch.nn as nn

class Autoencoder(nn.Module):
    def __init__(self, encoder_hidden_dims, decoder_hidden_dims, input_channels):
        super(Autoencoder, self).__init__()
        self.input_channels = input_channels
        
        # Encoder layers
        encoder_layers = []
        for i in range(len(encoder_hidden_dims)):
            if i == 0:
                encoder_layers.append(nn.Conv2d(input_channels, encoder_hidden_dims[i], kernel_size=3, stride=1, padding=1))
            else:
                encoder_layers.append(nn.BatchNorm2d(encoder_hidden_dims[i-1]))
                encoder_layers.append(nn.ReLU())
                encoder_layers.append(nn.Conv2d(encoder_hidden_dims[i-1], encoder_hidden_dims[i], kernel_size=3, stride=1, padding=1))
        self.encoder = nn.Sequential(*encoder_layers)
             
        # Decoder layers
        decoder_layers = []
        for i in range(len(decoder_hidden_dims)):
            if i == 0:
                decoder_layers.append(nn.Conv2d(encoder_hidden_dims[-1], decoder_hidden_dims[i], kernel_size=3, stride=1, padding=1))
            else:
                decoder_layers.append(nn.BatchNorm2d(decoder_hidden_dims[i-1]))
                decoder_layers.append(nn.ReLU())
                decoder_layers.append(nn.Conv2d(decoder_hidden_dims[i-1], decoder_hidden_dims[i], kernel_size=3, stride=1, padding=1))
        decoder_layers.append(nn.Conv2d(decoder_hidden_dims[-1], input_channels, kernel_size=3, stride=1, padding=1))  # Output layer
        self.decoder = nn.Sequential(*decoder_layers)

    def forward(self, x):
        x = self.encoder(x)
        x = x / x.norm(dim=(2, 3), keepdim=True)
        x = self.decoder(x)
        x = x / x.norm(dim=(2, 3), keepdim=True)
        return x

    def encode(self, x):
        x = self.encoder(x)
        x = x / x.norm(dim=(2, 3), keepdim=True)
        return x

    def decode(self, x):
        x = self.decoder(x)
        x = x / x.norm(dim=(2, 3), keepdim=True)
        return x
```
